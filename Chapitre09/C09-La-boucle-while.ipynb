**Boucle while**

```python
i = 1
while(i <= 3):
  print("Bienvenue à Python")
  i = i+1

print("Fin de l'exécution")
```

**Exemple de boucle while avec suivi pas à pas**

```python
i = 1
while(i <= 5):
  print(i)
  i = i+1

print("Fin de l'exécution")
```

**Boucles contrôlées par des sentinelles**

```python
prenom = ""
while(prenom!= "Daniel"):
  prenom = input("Saisissez un prénom : ")

print("Bienvenue")
```

**Déclaration break**

```python
while(True):
  prefecture = input("La préfecture du département de la Moselle en Lorraine, quel est-il ?")
  if(prefecture == "Metz"):
    print("Félicitations, vous avez gagné le jeu-concours!")
    break

print("Fin de l'exécution")
```
